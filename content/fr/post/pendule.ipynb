```julia
using Plots
```

```julia
l = 1e-2; g = 9.81
t = 0; θ = deg2rad(30); ω = 0
Δt = 1e-6
solθ = []
iter = 0
while t < 1
    dω = - g / l * sin(θ); dθ = ω
    ω += Δt * dω; θ += Δt * dθ
    t += Δt
    if iter % 1000 == 0
        append!(solθ, [θ])
    end
    iter += 1
end
```

```julia
p = plot(range(0, 1, length=length(solθ)), solθ, color="black", label="")
xlims!(0, 1)
xlabel!("t")
ylabel!("θ")
savefig(p, "../../../static/img/ode_pde/pendule_01.svg")
```

```julia
p = plot()
for dt in [1e-4, 1e-3, 2e-3, 3e-3, 4e-3, 5e-3]
    l = 1e-2; g = 9.81
    t = 0; θ = deg2rad(30); ω = 0
    Δt = dt
    solθ = []
    iter = 0
    while t < 1
        dω = - g / l * sin(θ); dθ = ω
        ω += Δt * dω; θ += Δt * dθ
        t += Δt
        append!(solθ, [θ])
        iter += 1
    end

    plot!(p, range(0, 1, length=length(solθ)), solθ, label="Δt= $(dt)")
end
xlims!(0, 1)
ylims!(-3.5, 2.5)
xlabel!("t")
ylabel!("θ")
plot!(p, legend=:bottomleft)
savefig(p, "../../../static/img/ode_pde/pendule_02.svg")
```
